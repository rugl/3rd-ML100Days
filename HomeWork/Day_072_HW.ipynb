# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

x = np.arange(-10,10,0.01)
```

```python
def sigmoid(x):
    return (1 / (1 + np.exp(-x)))

plt.plot(x, sigmoid(x))
plt.grid(color='b',linewidth='0.2' ,linestyle='--')
plt.title('Sigmoid Function')
```

```python
def dsigmoid(x):
    return sigmoid(x) * (1 - sigmoid(x))

plt.plot(x, dsigmoid(x))
plt.grid(color='b',linewidth='0.2' ,linestyle='--')
plt.title('Derivative Sigmoid Function')
```

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

plt.plot(x, tanh(x))
plt.grid(color='b',linewidth='0.2' ,linestyle='--')
plt.title('Tanh Function')
```

```python
def dtanh(x):
    return 1 - np.square(tanh(x))

plt.plot(x, dtanh(x))
plt.grid(color='b',linewidth='0.2' ,linestyle='--')
plt.title('Derivative Tanh Function')
```

```python

def reLU(x):
    return list(map(lambda x:x if x > 0 else 0, x))

plt.plot(x, reLU(x))
plt.grid(color='b',linewidth='0.2' ,linestyle='--')
plt.title('ReLU Function')
```

```python

def dReLU(x):
    return list(map(lambda x : 1 if x > 0 else 0, x))

plt.plot(x, dReLU(x))
plt.grid(color='b',linewidth='0.2' ,linestyle='--')
plt.title('Derivative ReLU Function')
```

```python
def leakyReLU(x):
    return list(map(lambda x:x if x > 0 else 0.1 * x, x))

plt.plot(x, leakyReLU(x))
plt.grid(color='b',linewidth='0.2' ,linestyle='--')
plt.title('Leaky ReLU Function')
```

```python
def dLeakyReLU(x):
    return list(map(lambda x: 1 if x > 0 else 0.1, x))

plt.plot(x, dLeakyReLU(x))
plt.grid(color='b',linewidth='0.2' ,linestyle='--')
plt.title('Derivative Leaky ReLU Function')
```
